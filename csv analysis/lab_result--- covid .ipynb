```python
COMPONENT_ID_covid = ['20126','202303','20240','20241','20242','20243','20244','20245','20264','20274','20275','20276','20277','20278','20279','20280','20283','20285','20286','20287','20288','20289','20290','20291','20292','20293','20298','20300','20301','20303','20308','20346','20347','20364','20365','20369','20379','20380','20381','20389','20390','20391','20392','20460','20463','20464','20465','20466','20469','20640','20647','20652','20655','20670']

lab_covid = lab[lab.COMPONENT_ID.isin(COMPONENT_ID_covid)]
```

```python
lab_covid=lab_covid.copy()

lab_covid['RESULT_DATE']=pd.to_datetime(lab_covid.loc[:,'RESULT_DATE'], format='%m/%d/%Y')
lab_covid['SPECIMEN_COLLECTED_DATE']=pd.to_datetime(lab_covid.loc[:,'SPECIMEN_COLLECTED_DATE'], format='%m/%d/%Y')
lab_covid['SPECIMEN_RECEIVED_DATE']=pd.to_datetime(lab_covid.loc[:,'SPECIMEN_RECEIVED_DATE'], format='%m/%d/%Y')
```

```python
num_ID = lab_covid.STUDY_ID.nunique()
print('the number of unique study ID: ', num_ID)

```

```python
# extract covid patient's STUDY_ID and its corresponding RESULT_DATE 
#----  for the purpose of extraction covid patient in Encounter table

ID_date_dic = {}
n = lab_covid.shape[0]
for i in range(n):
    ID = lab_covid.iloc[i,0]
    result_date = lab_covid.iloc[i,4]
    ID_date_dic[ID] = ID_date_dic.get(ID,[])+[result_date]
```

```python
# 
ID_date_dic
```

```python
covid_patient_lab = lab_covid.STUDY_ID.values

```

```python
covid_patient_lab

```

# encounter  +  lab_covid----encounter_covid_lab

```python
encounter_covid_lab_tem = Encounter[Encounter.STUDY_ID.isin(covid_patient_lab)]
encounter_covid_lab_tem =  encounter_covid_lab_tem.sort_values('START_DATE')
encounter_covid_lab_tem= encounter_covid_lab_tem.iloc[-900000:]
encounter_covid_lab_tem.head(10)
```

```python
ind =[]
n =  encounter_covid_lab_tem.shape[0]
count =  0 
id_tem = -1
for i in range(n):
    data = encounter_covid_lab_tem.iloc[i,:]
    ENC_TYPE = data.ENC_TYPE
    STUDY_ID = data.STUDY_ID
    
    if ENC_TYPE == 'LAB RESULTS':
        count+=1
        s_date = data.START_DATE
        init_date = min(ID_date_dic[STUDY_ID])
        id_tem = STUDY_ID
        if s_date >=init_date :
            ind.append(i)
    elif i>0 and STUDY_ID==id_tem :
        encounter_covid_lab.append(data)
        ind.append(i)
    else:
        continue 
        
    if count == -1:
        break 
        
ind
```

```python
len(ind)
```

```python

encounter_covid_lab=encounter_covid_lab_tem.iloc[ind,:]  

# Remover column header and index
# save the result
# encounter_covid_lab stores all of records of covid patient
# for each patient, the initial record is the one getting LAB Results about COVID
# for this patient, after inital LAB result result, all of other record appearing after the initial will follow

encounter_covid_lab.to_csv("encounter_covid_lab.csv",index=False)
```
