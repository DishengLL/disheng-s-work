## Death Day

```python
null_in_death=DEMOGRAPHICS.DEATH_DATE.isnull().sum()
print("The number of death records:", DEMOGRAPHICS.shape[0]-null_in_death)
```

```python
17511/DEMOGRAPHICS.shape[0]
```

```python
## death table
DEMOGRAPHICS_death = DEMOGRAPHICS[DEMOGRAPHICS.DEATH_DATE.notna()]
DEMOGRAPHICS_death
```

```python
DEMOGRAPHICS[DEMOGRAPHICS.STUDY_ID==5220436245].head()
```

Here, I found that once a patient dies, all previous records on that patient are updated.  
Thus, a person may have several death records.   
In this dataset, in order to analyze deaths within 7 days, we only need to use the latest `ED_date`.

### Death exploration ---die within 7 days

```python
index=(DEMOGRAPHICS_death.DEATH_DATE-DEMOGRAPHICS_death.ED_DATE).dt.days<7 

death_within_7=DEMOGRAPHICS_death[index].copy()
death_within_7['within'] = list((DEMOGRAPHICS_death.DEATH_DATE-DEMOGRAPHICS_death.ED_DATE).dt.days.values[(DEMOGRAPHICS_death.DEATH_DATE-DEMOGRAPHICS_death.ED_DATE).dt.days<7])

death_within_7 = death_within_7[death_within_7.within>=0]
death_within_7 
```

```python
death_within_7.info()
```

```python
death_within_7.nunique()
```

```python
sns.displot(data= death_within_7,x= "AGE",aspect=2,kde = True,hue = "GENDER")
plt.title('age distribution of patients died within 7 days')
plt.show()
```

```python
death_within_7.DEATH_DATE.value_counts().sort_values()
```

```python
plt.figure(figsize=(100, 10))
tem = pd.DataFrame()
tem['date'] = death_within_7.DEATH_DATE.value_counts().sort_index().index
tem['count'] = list(death_within_7.DEATH_DATE.value_counts().sort_index())
sns.pointplot(x="date", y="count",data=tem)
plt.xticks(rotation = 90)
plt.grid()
plt.title('Death counts in Time Series---within 7 days')
plt.show()
```

### Death exploration --- die within 1 month (30 days)

```python
index=(DEMOGRAPHICS_death.DEATH_DATE-DEMOGRAPHICS_death.ED_DATE).dt.days<30 

death_within_30=DEMOGRAPHICS_death[index].copy()
death_within_30['within'] = list((DEMOGRAPHICS_death.DEATH_DATE-DEMOGRAPHICS_death.ED_DATE).dt.days.values[(DEMOGRAPHICS_death.DEATH_DATE-DEMOGRAPHICS_death.ED_DATE).dt.days<30])

death_within_30 = death_within_30[death_within_30.within>=0]
death_within_30
```

```python
death_within_30.info()
```

```python
death_within_30.nunique()
```

```python
sns.displot(data= death_within_30,x= "AGE",aspect=2,kde = True,hue="GENDER")
plt.title('age distribution of patients died within 30 days')
plt.vlines(60,0,200,linestyles ="dashed", colors ="r")
plt.vlines(79,0,230,linestyles ="dashed", colors ="r")
plt.show()
```

```python
death_within_30.DEATH_DATE.value_counts().sort_values()
```

```python
plt.figure(figsize=(300, 10))
tem = pd.DataFrame()
tem['date'] = death_within_30.DEATH_DATE.value_counts().sort_index().index
tem['count'] = list(death_within_30.DEATH_DATE.value_counts().sort_index())
sns.pointplot(x="date", y="count",data=tem)
plt.xticks(rotation = 90)
plt.grid()
plt.title('Death counts in Time Series---within 30 days')
plt.show()
```
